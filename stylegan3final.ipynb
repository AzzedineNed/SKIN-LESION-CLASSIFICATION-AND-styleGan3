## Finetuning a model

````python
batch_size = 16
batch_gpu_size = 8 # Batch size for GPU

dataset_path = '/content/drive/MyDrive/NEV_png_256x256.zip'
resume_from = '/content/drive/MyDrive/NEDJAI/stylegan3/results/00012-stylegan3-t-NEV-gpus1-batch16-gamma50/network-snapshot-000032.pkl'  #Can be the link from nvidia
#resume_from ='/content/drive/MyDrive/NEDJAI/stylegan3/models/network-a657-r-512-148863.pkl'


gamma_value = 50.0 #Adjustable R1 regularization weight
snapshot_count = 8
````

````python

# Fine-tune StyleGAN3-R for MetFaces-U using 1 GPU, starting from the pre-trained FFHQ-U pickle.
!python /content/drive/MyDrive/NEDJAI/stylegan3/train.py --outdir=/content/drive/MyDrive/NEDJAI/stylegan3/results --cfg=stylegan3-t \
--data=$dataset_path --gpus=1 --batch=$batch_size --batch-gpu=$batch_gpu_size --gamma=$gamma_value --kimg=1000 \
--snap=$snapshot_count --workers=2  --metrics='pr50k3_full'

````
